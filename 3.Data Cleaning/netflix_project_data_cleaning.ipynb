## Handling Missing Values
### Fill the missing values and drop the remaining ones

> The 'director', 'cast', and 'country' columns contained missing values, which represent important descriptive data. Instead of dropping rows, which would reduce the dataset size and potentially exclude meaningful insights, I chose to fill the missing values. For director and cast, I used "Unknown" and "Not Available" as placeholders to retain all rows while explicitly acknowledging gaps. For country, I filled missing values with "Unknown" to maintain consistency for geographical analysis.

```python
# Filling missing values

df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['country'].fillna('Unknown', inplace=True)
```

```python
df.info()
```

```python
# Drop rest of the missing values(in date_added, rating, duration)
#as they are not so much values

df.dropna(inplace=True)
```

```python
def summary(df):
    summ = pd.DataFrame(df.dtypes, columns=['data type'])
    summ['#missing'] = df.isnull().sum().values
    summ['Duplicate'] = df.duplicated().sum()
    summ['#unique'] = df.nunique().values
    desc = pd.DataFrame(df.describe(include='all').transpose())
    summ['min'] = desc['min'].values
    summ['max'] = desc['max'].values
    summ['avg'] = desc['mean'].values
    summ['std dev'] = desc['std'].values
    summ['top value'] = desc['top'].values
    summ['Freq'] = desc['freq'].values

    return summ

summary(df).style.background_gradient()
```

```python
# Convert integers to months
df['month_added'] = df['month_added'].map(months_dict)
```
